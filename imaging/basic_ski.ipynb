```python

import sys
import subprocess
colab = False 

if "google.colab" in sys.modules:
    # Use Colab/Jupyter magic for installation
    colab = True
    %pip install -q scikit-image

import skimage as ski
import numpy as np

if colab:
    img = ski.io.imread("https://raw.githubusercontent.com/LOMA-AImotion/pdl_ss26/main/imaging/robos.jpg")
else: 
    img = ski.io.imread("robos.JPG")
    
print("Number of dimensions:", img.ndim)
print("Shape:", img.shape)

import matplotlib.pyplot as plt

def imshow_matplotlib(image, cmap=None, **kwargs):
    plt.figure()
    if image.ndim == 2: # Grayscale image has only width and height as dimensions
        plt.imshow(image, cmap=cmap or "gray", **kwargs)
    else:
        plt.imshow(image, **kwargs)
    plt.axis("off")
    plt.show()
    
imshow_matplotlib(img)

img_grey = ski.color.rgb2gray(img)
print("Number of dimensions:", img_grey.ndim)
print("Shape:", img_grey.shape)
imshow_matplotlib(img_grey)
```

```python
# Set pixel coordinates (x: column, y: row)

x, y = 800, 450  # change these values

# Validate bounds and print RGB
h, w = img.shape[:2]
if 0 <= x < w and 0 <= y < h:
    r, g, b = img[y, x]
    print(f"Pixel at (x={x}, y={y}) -> RGB: ({r}, {g}, {b})")
else:
    print(f"Coordinates out of bounds. Valid ranges: x in [0, {w-1}], y in [0, {h-1}]")  

# Mark pixel (x, y) and display its color information
if 0 <= x < w and 0 <= y < h:
    pixel = img[y, x]  # [R, G, B]
    r, g, b = pixel

    print(f"Pixel at (x={x}, y={y})")
    print(f"Original color value: {pixel}")
    print(f"R: {r}, G: {g}, B: {b}")

    plt.figure(figsize=(8, 6))
    plt.imshow(img)
    plt.scatter(x, y, c="red", s=1, marker="o")
    plt.title(f"Pixel ({x}, {y}) -> RGB=({r}, {g}, {b})")
    plt.axis("off")
    plt.show()
else:
    print(f"Coordinates out of bounds. Valid ranges: x in [0, {w-1}], y in [0, {h-1}]")


# Display the RGB color as a solid patch
color_patch = np.full((200, 200, 3), [r, g, b], dtype=np.uint8)

plt.figure(figsize=(3, 3))
plt.imshow(color_patch)
plt.title(f"RGB = ({r}, {g}, {b})")
plt.axis("off")
plt.show()
```
