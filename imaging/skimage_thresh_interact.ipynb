```python
import sys
colab = False 

if "google.colab" in sys.modules:
    # Use Colab/Jupyter magic for installation
    colab = True
    %pip install -q scikit-image

import skimage as ski
import numpy as np

if colab:
    img = ski.io.imread("https://raw.githubusercontent.com/LOMA-AImotion/pdl_ss26/main/imaging/thi.jpg")
else: 
    img = ski.io.imread("thi.JPG")
    
print("Number of dimensions:", img.ndim)
print("Shape:", img.shape)
print("Data type:", type(img))

import matplotlib.pyplot as plt

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import ipywidgets as widgets
from ipywidgets import interact

threshold_value = 80
def show_threshold(threshold_value=80):
    global img_thresholded
    img_thresholded = np.where(img.mean(axis=2) > threshold_value, 255, 0).astype(np.uint8)

    fig, axes = plt.subplots(1, 2, figsize=(12, 6))

    axes[0].imshow(img)
    axes[0].set_title("Original")
    axes[0].axis("off")

    axes[1].imshow(img_thresholded, cmap="gray")
    axes[1].set_title(f"Thresholded (> {threshold_value})")
    axes[1].axis("off")

    plt.tight_layout()
    plt.show()

interact(
    show_threshold,
    threshold_value=widgets.IntSlider(
        value=threshold_value, min=0, max=255, step=1, description="Threshold"
    )
);
```
